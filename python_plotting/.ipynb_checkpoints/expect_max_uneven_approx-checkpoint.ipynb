```python
%matplotlib inline  

import numpy as np
import math as m

import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as mcolors

rez=600
```

```python
#function to compute the ZTP distribution
def ztp(r,lam): 
    if r>0:
        p_ztp=lam**r/(np.exp(lam)-1)/m.factorial(r)
    else:
        p_ztp=0
    return p_ztp

#the mean of the ztp distribution
def ztp_avg(lam):
    return lam/(1-np.exp(-lam))
```

```python
#explore ZTP

r=np.arange(1,10)
lam=3

p_ztp=[]
for R in r:
    p_ztp.append(ztp(R,lam))

plt.plot(r,p_ztp)
plt.xlabel('$R$')
plt.ylabel('$\mathcal{Z}(\lambda,R)$')
plt.title('$\lambda=$'+str(lam))
print('average = ',ztp_avg(lam))
```

```python
#Calculating the multiplicity of multinomial combinations
#the number of words  found in the expansion (p1+p2+p3+...R)^n that have exactly k components. 
def p_Robs(n,R,alpha):
    g_k = np.zeros(R) #make an array to hold the probabilities of exactly k strains, where k\le R
    
    phi=(np.exp(alpha)-1)/(1-np.exp(-alpha*R)); #normalization for exponential sum

    #k needs to run from 1->R, p_{k=0} = 0
    for k in range(1,R+1,1):
        inc_exc = 0.0 #include exclude factor
        #i runs from 0 to k-1
        for i in range(k):
            #print(inc_exc)
            inc_exc += m.factorial(k)/m.factorial(k-i)/m.factorial(i)*(-1)**i*(k-i)**n     
        
        firstexp = k*(k-1)/2; #the sum of the first terms for the approx exponential
        
        unf = m.factorial(k)/(phi**k)*np.exp(-alpha*firstexp) #uneven function, the probability of a selection
        
        g_k[k-1] = m.factorial(R)/m.factorial(k)/m.factorial(R-k)*inc_exc*unf #multiplicity and probability
        
    return g_k/sum(g_k)

#calculates the conditional probability p(Robs|n,R)p(R) for many hosts
def dist_product(n,R_list,lam,al):
    p_combo=np.zeros([len(R_list),len(R_list)]) #matrix of zeros for conditional prob
    for i in range(len(R_list)):
        R=R_list[i]
        pp=p_Robs(n,R,al)*ztp(R,lam)
        p_combo[i,0:len(pp)]=pp #fill up this matrix

    return p_combo
```

```python
#visualize p(R,Robs), e.g. can't measure more strains than underlying, or more than n
n=7
R_list=np.arange(1,10)
lam=1
fig,axarr=plt.subplots(1,3,sharey=True,sharex=True,figsize=(7,3),dpi=rez)
for al in range(1,4):
    plt.subplot(1,3,al)
    p_combo=dist_product(n,R_list,lam,al)
    plt.imshow(np.log10(p_combo),interpolation='none',vmin=-5,vmax=0)
    plt.axvline(n-1,color='k',ls='--')
    plt.annotate('n='+str(n),[n-0.5,1])
    
    plt.xticks(np.arange(10),np.arange(1,10))
    plt.yticks(np.arange(10),np.arange(1,10))
    plt.ylabel('$R$')
    plt.xlabel('$R^{obs}$')
    plt.title(r'$\alpha$= '+str(al))

cbar=plt.colorbar()
#cbar=plt.colorbar(ax=axarr.ravel().tolist())
cbar.set_label('$\log_{10} \, p(R,R^{obs})$')#, rotation=270)
plt.tight_layout()

plt.savefig('figures/uneven_approx_probs.pdf')
```

```python
#simulate UNeven data
def simulate_data(lambda_sim,alpha,N_sim,n_sim):
    
    trial = np.random.poisson(lambda_sim,[N_sim*100]) #workaround to avoid drawing from ZTP
    trial = trial[trial>0] #only take participants with Robs>0
    true_trial = trial[0:N_sim] #just use the sims that matter
    
    Robs=np.zeros(10) #Robs up to max value, 10 strains for example
    
    #loop through patients
    for i in range(len(true_trial)):
        R=true_trial[i]
        
        r = np.arange(1,R+1)
        p_j = np.exp(-alpha*r)/sum(np.exp(-alpha*r))
        
        samples = np.random.multinomial(n_sim,p_j)
        Robs[sum(samples>0)-1]+=1 #add a count to the histogram
    
    return Robs

#function that computes the likelihood of the data given lambda
def likelihood(counts,lam,n,alpha):
    r=np.arange(1,len(counts)+1) #truncate the possible maximum richness with the data
    C=sum(dist_product(n,r,lam,al)) #the marginal
    Lik=C/np.sum(C) #the conditional probability we want
    
    #print(np.log(Lik))
    logLik = np.nansum(counts*np.log(Lik)) #compute the log likelihood given data and probabilty
    
    return logLik

#compute the maximum of the likelihood
def max_likelihood(lam_list,al_list,counts,n):
    L_list=np.zeros(len(lam_list))
    #L_list=np.zeros([len(lam_list),len(al_list)])
    linx=0
    for i in range(len(lam_list)):
        #for j in range(len(al_list)):
        L_list[i]=likelihood(counts,lam_list[i],n,al)        
    maxL_index=np.argsort(L_list)[-1]
    
    maxL=L_list[maxL_index]
    max_lam=lam_list[maxL_index]
    return L_list,max_lam,maxL
```

```python
#check simulated data for different n and N:

lam=1
N=10000
n=100
alpha=3

dat=simulate_data(lam,alpha,N,n)

r=np.arange(1,len(dat)+1) #list of richnesses we can observe

p_ztp=[]
for R in r:
    p_ztp.append(ztp(R,lam))

plt.plot(r,dat/N,'o')
plt.plot(r,p_ztp)
plt.xlabel('observed richness')
plt.ylabel('probability of \n observed richness')

plt.legend(['sim','theory'])
```

```python
#uneven example, demonstrate the likelihood curves

lam_test=2
al_test=2

N=1000
n=3

counts=simulate_data(lam_test,al_test,N,n) #simulate data
r=np.arange(1,len(counts)+1)
naive=sum(counts*r)/sum(counts)

lam_list=np.logspace(-1,2,1000) #from Ravg=1.05 to 5

p_ztp=[]
for R in r:
    p_ztp.append(ztp(R,lam)*N)
        
l=[]
xerrz=np.zeros([2,1])
plt.figure(figsize=(5,5))
for al in np.linspace(1,3,11):
    
    L_list,max_lam,maxL = max_likelihood(lam_list,al,counts,n)

    p=plt.semilogx(lam_list,L_list,alpha=0.5)
    plt.xlabel('ZTP parameter ($\lambda$)')
    plt.ylabel('log likelihood ($\log \mathcal{L}$)')

    CI = lam_list[np.argwhere(np.diff(np.sign(L_list - maxL + 2)) != 0).reshape(-1)]

    if len(CI)>1:
        xerrz[0]=max_lam-2*CI[0]
        xerrz[1]=2*CI[1]-max_lam
    else:
        xerrz[0]=max_lam-CI[0]
        xerrz[1]=lam_list[-1]-max_lam

    plt.errorbar(max_lam, max(L_list)+25, xerr=xerrz, fmt='o',color=p[0].get_color())

    l.append(r'$\alpha=$'+str(al))

plt.ylim([-800,max(L_list)+100])
plt.axvline(lam_test,color='k',ls='--')
plt.axvline(naive,color='gray',ls=':')
l.append('true')
l.append('naive')
plt.legend(l,fontsize=8)
#plt.axhline(maxL-2,color='k',ls='--')

plt.tight_layout()

```

```python
#calculate the psuedo evenness given always could be Rmax=10 
def evenness(al):
    x=np.arange(1,10)
    p_x=np.exp(-al*x)/np.sum(np.exp(-al*x))

    S = -np.sum(p_x*np.log(p_x)) 
    E = S/np.log(max(x))

    return E

```

```python
#uneven example, infer both??

lam_test=2
al_test=2

N=1000
n=10

counts=simulate_data(lam_test,al_test,N,n) #simulate data
r=np.arange(1,len(counts)+1)
naive=sum(counts*r)/sum(counts)

lam_list=np.logspace(-1,1,100) #from Ravg=1.05 to 5
Ravg=ztp_avg(lam_list) #the vector of average richnesses
al_list=np.linspace(1,3,100) #superinfectino parameter range

p_ztp=[]
for R in r:
    p_ztp.append(ztp(R,lam)*N)
        
L_al_list=[]
for al in al_list:
    L_list,max_lam,maxL = max_likelihood(lam_list,al,counts,n)
    L_al_list.append(L_list)

mZ=np.array(L_al_list) #make it an array

best_fit = np.where(mZ==np.max(mZ)) #location of best fit

best_R = ztp_avg(lam_list[best_fit[1][0]])
best_al = al_list[best_fit[0][0]]

errz=[(best_R-ztp_avg(lam_test))/ztp_avg(lam_test), (best_al-al_test)/al_test]
```

```python
for i in range(len(L_al_list)):
    plt.plot(lam_list,L_al_list[i],color=[0,0,i/len(L_al_list)],alpha=0.5)
#plt.colorbar()
```

```python
#plot simple figure for mutual estimation

plt.figure(figsize=(5,3))

E_list=[evenness(al_list[i]) for i in range(len(al_list))]

levelz=np.linspace(-500,0,11)
plt.contourf(Ravg,E_list,mZ-np.max(mZ),levels=levelz)
cbar=plt.colorbar()
s1=plt.scatter(ztp_avg(lam_test),evenness(al_test),edgecolors='k',marker='o',facecolors='none') #true answer
s2=plt.scatter(best_R,evenness(best_al),c='k',marker='o')
cbar.set_label('log decrease from \n maximum log likelihood')
plt.xlabel('average richness')
plt.ylabel('evenness')
plt.legend([s1,s2],['truth','MLE'])

plt.tight_layout()
```
