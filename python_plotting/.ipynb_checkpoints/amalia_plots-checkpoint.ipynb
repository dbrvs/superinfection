```python
%matplotlib inline  

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import pandas as pd

plt.style.use('ggplot')
```

```python
##############################
#make line plots n vs error as n*N varies, for given error metric, and for all alphas
#answers the question of bias with n 

am_sim = pd.read_csv('estlambda_500pers.csv')

n_list = list(am_sim.nsamp.unique())
alpha_list = list(am_sim.alpha.unique())
lambda_list = list(am_sim.lamsim.unique())

f1, axarr1 = plt.subplots(1,len(alpha_list),sharey=True,sharex=True,figsize=(6,1.5),dpi=600)
f2, axarr2 = plt.subplots(1,len(alpha_list),sharey=True,sharex=True,figsize=(6,1.5),dpi=600)

px=0
for al in alpha_list:
    temp1_df = am_sim[am_sim['alpha']==al]
    
    for lam in lambda_list:
        temp2_df = temp1_df[temp1_df['lamsim']==lam]
        
        axarr1[px].plot(n_list,temp2_df['stddevlamest'])#/lam)
        axarr2[px].plot(n_list,temp2_df['pbiaslamest'])
    
    axarr1[px].set_title(r'$\alpha=$'+ str(al),fontsize=12)
    #axarr2[px].set_xlabel('$n$')
    #axarr1[px].set_ylim([0,0.15])
    #axarr2[px].set_ylim([-100,25])

    px+=1
    
axarr1[0].set_ylabel('std dev $\lambda$')
axarr2[0].set_ylabel('bias $\lambda$ (%)')
axarr1[0].set_xticks(range(min(n_list),max(n_list)+1))
axarr1[0].set_xticklabels([])
axarr2[0].set_xticks(range(min(n_list),max(n_list)+1))
axarr2[0].set_xticklabels([])

#axarr1[0].legend(['$p_\mathcal{S}$=0.01','0.02','0.05','0.10','0.20','0.30'],loc=1,fontsize=8)

f1.tight_layout()
f1.savefig('fixed_N_stddev.pdf')  # No warning now
f2.tight_layout()
f2.savefig('fixed_N_bias.pdf')  # No warning now

```

```python
#the mean of the ztp distribution
def ztp_avg(lam):
    return lam/(1-np.exp(-lam))
```

```python
##############################
am_sim = pd.read_csv('estlambda_fixedsize_more.csv')

n_list = list(am_sim.nsamp.unique())
N_list = list(am_sim.npers.unique())
alpha_list = list(am_sim.alpha.unique())
lambda_list = list(am_sim.lamsim.unique())

f1, axarr = plt.subplots(len(alpha_list),len(lambda_list),sharey=True,sharex=True,figsize=(10,5),dpi=600)

normalize=mcolors.Normalize(-10,10)
#normalize=mcolors.Normalize(min(am_sim.pbiaslamest),max(am_sim.pbiaslamest))
colormap=cm.jet

alx=0
for al in alpha_list:
    lx=0
    for lam in lambda_list:
        df = am_sim.query('(lamsim==' + str(lam) +') & (alpha==' + str(al) + ')')
        axarr[alx][lx].scatter(df.nsamp,df.npers,color=colormap(normalize(df.pbiaslamest)),
                               s=df.stddevlamest*2e3,alpha=0.7)#,edgecolors='k')
        if alx==0:
            axarr[alx][lx].set_title(r'$\langle R \rangle=$'+str(np.round(ztp_avg(lam),3)),fontsize=10)
        axarr[alx][lx].set_axis_bgcolor('silver')
        lx+=1
    axarr[alx][0].set_ylabel(r'$\alpha=$'+str(al),fontsize=12)
    alx+=1

axarr[0][0].set_xlim([1.5,6.5])
axarr[0][0].set_ylim([-100,1100])
f1.tight_layout()

# setup the colorbar
scalarmappaple=cm.ScalarMappable(norm=normalize, cmap=colormap)
scalarmappaple.set_array(am_sim.pbiaslamest)
cbar=plt.colorbar(scalarmappaple,ax=axarr.ravel().tolist())
#cbar.set_clim([-8,8])
#cbar.set_ticks(nlist) # add the labels
#cbar.set_ticklabels(nlist) # add the labels
#cbar.ax.set_title('% bias',fontsize=10)

plt.savefig('amalia_bubbles.pdf')
```

```python
############################################################
#make line plots n vs error for several different n*N stays constant
#separate by n*N

am_sim = pd.read_csv('estlambda_fixedsize_more.csv')

n_list = am_sim.nsamp.unique()
N_list = am_sim.npers.unique()
alpha_list = am_sim.alpha.unique()
lambda_list = am_sim.lamsim.unique()
nN=am_sim.totsamp.unique()

for inN in range(len(nN)):

    f, axarr = plt.subplots(2,len(alpha_list),sharey=False,sharex=True,figsize=(7,4),dpi=600)

    am_sim_nN = am_sim[am_sim.totsamp == nN[inN]]

    for ia in range(len(alpha_list)):

        for il in range(len(lambda_list)):
            
            lam=lambda_list[il]
            al=alpha_list[ia]
            
            df = am_sim_nN.query('(lamsim==' + str(lam) +') & (alpha==' + str(al) + ')' )

            axarr[0][ia].plot(n_list,df['stddevlamest'])
            axarr[1][ia].plot(n_list,df['pbiaslamest'])

        axarr[0][ia].set_title(r'$\alpha=$'+ str(al),fontsize=12)
        axarr[1][ia].set_ylim([-5,35])
        axarr[0][ia].set_ylim([0,0.3])
    
    axarr[0][0].set_ylabel('estimate precision \n (std dev $\lambda$)')
    axarr[1][0].set_ylabel('estimate accuracy \n (% bias $\lambda$)')
    axarr[0][0].set_xticks(range(min(n_list),max(n_list)+1))

    axarr[1][2].text(3.5,25,r'$n\times N = $' + str(nN[inN]))

    axarr[1][1].set_xlabel('samples per ppt ($n$)')
    if inN==len(nN)-1:
        axarr[0][0].legend(['$p_\mathcal{S}$=0.01','0.02','0.05','0.10','0.20','0.30'],loc=2,fontsize=8)

    f.tight_layout()
    f.savefig('nN='+str(nN)+'.pdf')  # No warning now
```
